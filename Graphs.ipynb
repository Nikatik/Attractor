```python
from matplotlib import pyplot as plt
from tqdm import tqdm
import subprocess
from subprocess import Popen, PIPE, STDOUT

p = subprocess.Popen(f'make', shell=True, stdout=PIPE, stderr=STDOUT)
p.wait()
```

```python
plt.figure(figsize=(20, 15), dpi=300)
plt.xlabel('x1')
plt.ylabel('x3')
R_list = [1.4, 1.9, 2.5, 5, 10, 13.929, 20]
for R in tqdm(R_list):
    x = []
    y = []
    c = []
    p = subprocess.Popen(f'./main 17 {R}', shell=True, stdout=PIPE, stderr=STDOUT)
    p.wait()
    with open(f'./trajectory.txt') as f:
            x = []
            y = []
            cf = f.read().replace(" ", "").splitlines()
            for i in range(1, len(cf) + 1):
                cf[len(cf) - i] = cf[len(cf) - i].split('|')
                x.append(-float(cf[len(cf)-i][1]))
                y.append(float(cf[len(cf)-i][3]))
            for i in range(0, len(cf)):
                x.append(float(cf[i][1]))
                y.append(float(cf[i][3]))
            plt.plot(x,y, linewidth=1., label=f'R = {str(R)}')    
plt.legend()
plt.grid(True)
plt.savefig('plot.png', transparent=False)
plt.show()
```
